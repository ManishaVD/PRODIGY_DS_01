```python
import matplotlib.pyplot as plt
```

```python
# Bar Chart for Income Distributions by Regions
income_by_region = pd.crosstab(df2.IncomeGroup, df2.Region)
plt.figure(figsize=(10, 6))
income_by_region.plot(kind='bar')
plt.xlabel("Income Group")
plt.ylabel("Number of Countries in the Regions")
plt.title("Income Distribution by Regions")
plt.legend(title='Regions')
plt.xticks(rotation=45)
plt.show()
```

 Interpretation for the above graph: The graph highlights the varying income levels across different regions of the world, with some regions having a higher concentration of high-income or low-income countries, while others have a more diverse income distribution.

```python
#top 10 countries based on population in the year 1973
df_sorted1 = df1.sort_values(by='1973', ascending = False).head(10)
plt.figure(figsize=(10,6))
plt.barh(df_sorted1['Country Name'], df_sorted1['1973'], color = 'Red')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 1973')
plt.gca().invert_yaxis() 
plt.show()
```

 Interpretation for the above graph: The graph provides a visual comparison of the relative population sizes of these different country groups and income levels in 1973, with the world population being the largest and serving as a reference point.

```python
#bottom 10 country based on population in the year 1973
df_sorted3 = df1.sort_values(by='1973', ascending=False).tail(10)
plt.figure(figsize=(10, 6))
plt.barh(df_sorted3['Country Name'], df_sorted3['1973'], color='yellow')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 1973')
plt.gca().invert_yaxis()
plt.show()
```

 Interpretation for the above graph:The graph effectively visualizes the countries with the lowest populations globally in 1973, providing a clear ranking and comparative view of these small nations or territories.Palau had the smallest population among these countries, followed by Cayman Islands and British Virgin Islands. The graph displays the relative population sizes through the lengths of the yellow bars, allowing for easy comparison across these countries with very low populations in 1973.

```python
#top 10 countries based on population in the year 2022
df_sorted1 = df1.sort_values(by='2022', ascending = False).head(10)
plt.figure(figsize=(10,6))
plt.barh(df_sorted1['Country Name'], df_sorted1['2022'], color = 'purple')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 2022')
plt.gca().invert_yaxis() 
plt.show()
```

 Interpretation for the above graph: The graph provides a visual representation of the relative population sizes of different country groups in the year 2022, ranked from highest to lowest, based on income levels, demographic characteristics, or regional affiliations.The bars extend horizontally, with their lengths corresponding to the population size of each country or group.The top entry is labeled "World," indicating the total global population. The next entries represent different income and regional groups defined by the World Bank or other international organizations, such as "IDA & IBRD total" (likely referring to countries eligible for assistance from the International Development Association and International Bank for Reconstruction and Development), "Low & middle income," "Middle income," "IBRD only," and so on.

```python
#bottom 10 country based on population in the year 2022
df_sorted3 = df1.sort_values(by='2022', ascending=False).tail(10)
plt.figure(figsize=(10, 6))
plt.barh(df_sorted3['Country Name'], df_sorted3['2022'], color='darkblue')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 2022')
plt.gca().invert_yaxis()
plt.show()
```

 Interpretation for the above graph: The graph provides a visual representation of the relative population sizes of different country groups in the year 2022.The graph uses a horizontal bar chart format, where the length of the bars corresponds to the population size of each country. It allows for an easy comparison of the relative population sizes among these least populous nations in the world.Liechtenstein has the smallest population among the countries shown, followed by Monaco, San Marino, Gibraltar, and St. Martin (French part). The other countries in the bottom 10 by population are the British Virgin Islands, Palau, Nauru, and Tuvalu.

```python
# Histogram for population distribution in India from 1960-2022
data = df1[df1['Country Code'] == 'IND']
years = [str(year) for year in range(1960, 2023)]  
population = data[years].values.flatten()
plt.figure(figsize=(10, 6))
plt.hist(population, bins=15, edgecolor = 'black', color = 'pink')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution in India (1960-2022)')
plt.show()
```

 Interpretation for the above graph: The graph illustrates the population growth pattern in India, starting from a lower base and progressively increasing over the decades, with the majority of the population now concentrated in the higher ranges.The population distribution starts with a high frequency around 0.4-0.6 billion in the early years (likely representing the 1960s and 1970s), then gradually decreases in frequency as the population increases, before peaking again at around 1.4 billion in the most recent years (likely representing the present day or near future).
